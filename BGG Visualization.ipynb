```python
genre1 = []
genre2 = []
genre3 = []
genreRec = []

for index, row in df.iterrows():
    genreList = [item.strip(" ").strip("''") for item in row['genres'].strip('[]').split(',')]
    if len(genreList) == 1:
        genre1.append(genreList[0])
        genre2.append(None)
        genre3.append(None)
    elif len(genreList) == 2:
        genre1.append(genreList[0])
        genre2.append(genreList[1])
        genre3.append(None)
    elif len(genreList) == 3:
        genre1.append(genreList[0])
        genre2.append(genreList[1])
        genre3.append(genreList[2])
    else:
        genre1.append(genreList[0])
        genre2.append(None)
        genre3.append(None)
```

```python
for genreList in [genre1, genre2, genre3]:
    for index, el in enumerate(genreList):
        if el == "" or el == None:
            genreList[index] = "None"
```

```python
df.insert(4, "genre1", genre1, True)
df.insert(5, 'genre2', genre2, True)
df.insert(6, 'genre3', genre3, True)
```
